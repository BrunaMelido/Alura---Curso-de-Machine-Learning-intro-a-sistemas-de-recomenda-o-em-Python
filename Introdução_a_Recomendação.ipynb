```python
import pandas as pd
filmes = pd.read_csv("movies.csv")
filmes.columns = ["FilmeId", "Titulo", "Generos"]
filmes = filmes.set_index("FilmeId")
filmes.head()
```

```python
notas = pd.read_csv("ratings.csv")
notas.columns = ["UsuarioId", "FilmeId", "Nota", "Momento"]
notas.head()
```

```python
notas.describe()
```

#Primeira Tentativa de Recomendação: heurística de total de votos

```python
total_de_votos = notas["FilmeId"].value_counts()
total_de_votos.head()
```

```python
filmes["total_de_votos"] = total_de_votos
filmes.head()
```

```python
filmes.sort_values("total_de_votos", ascending = False).head()
```

```python
notas_medias = notas.groupby("FilmeId").mean()["Nota"]
notas_medias.head()
```

```python
filmes["notas_media"] = notas_medias
filmes.sort_values("total_de_votos", ascending = False).head(10)
```

# Uma Segunda heurística

```python
filmes.sort_values("notas_media", ascending = False).head(10)
```

```python
filmes.query("total_de_votos >= 10").sort_values("notas_media", ascending = False).head(10)
```

```python
filmes_com_mais_de_50_votos = filmes.query("total_de_votos >= 50")
filmes_com_mais_de_50_votos.sort_values("notas_media", ascending = False).head(10)
```
